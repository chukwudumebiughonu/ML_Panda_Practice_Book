## Creating Data from existing data

```python
# Columns from series
seatColumn = pd.Series([4,4,4,4,])

#New column called seats
car_sales['seats'] = seatColumn
```

```python
car_sales
```

```python
car_sales['seats'].fillna(5, inplace = True)
```

```python
car_sales
```

```python
# Creating a column from python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales['Fuel per 100km'] = fuel_economy
```

```python
car_sales
```

```python
car_sales['Fuel per 100km'].plot();
```

```python
car_sales['Fuel per 100km'].hist();
```

```python
car_sales['Total used fuel'] = car_sales['Odometer (KM)']/100 * car_sales['Fuel per 100km']
```

```python
car_sales
```

```python
car_sales['Total used fuel'].plot();
```

```python
car_sales['Total used fuel'].hist()
```

```python
# Creating a colum from a single python column
car_sales['Number of wheel'] = 4
car_sales
```
